```python
import pandas as pd
import numpy as np
data = pd.read_csv('hotel_booking.csv')
```

```python
data
```

```python
columns_to_remove = [
    'lead_time',
    'arrival_date_week_number',
    'arrival_date_day_of_month',
    'stays_in_weekend_nights',
    'stays_in_week_nights',
    'meal',
    'market_segment',
    'distribution_channel',
    'is_repeated_guest',
    'previous_cancellations',
    'previous_bookings_not_canceled',
    'booking_changes',
    'deposit_type',
    'agent',
    'company',
    'days_in_waiting_list',
    'customer_type',
    'adr',
    'required_car_parking_spaces',
    'total_of_special_requests',
    'name',
    'email',
    'phone-number',
    'credit_card'
]

# Remove the specified columns
df_cleaned = data.drop(columns=columns_to_remove)

# Save the cleaned DataFrame to a new CSV file
cleaned_file_path = "hotel_booking_cleaned.csv"
df_cleaned.to_csv(cleaned_file_path, index=False)

cleaned_file_path
```

```python
dataset = pd.read_csv('hotel_booking_cleaned.csv')
dataset
```

```python
dataset['room_status'] = np.where(dataset['reserved_room_type'] == dataset['assigned_room_type'],"Desired","Un-Desired")
```

```python
dataset
```

```python
conditions = [
    (dataset["adults"] == 2) & (dataset["children"] == 0) & (dataset["babies"] == 0),
    (dataset["adults"] == 1) & (dataset["children"] == 0) & (dataset["babies"] == 0)
]

# Define outputs
choices = ["Couples", "Single"]
```

```python
dataset["guest_type"] = np.select(conditions, choices, default="Family")
```

```python
dataset
```

```python
dataset.to_csv("hotel_booking_cleaned.csv",index=False)
```
