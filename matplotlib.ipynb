# Matplotlib Practice

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = [98,67,85,95,88]
x = ["Part1","Part2","Part3","Part4","Part5"]
color = ["red","blue","green","yellow","orange"]
plt.bar(x,y,color = color,edgecolor="black")
plt.xlabel("Parts of Harry Potter",fontsize=20)
plt.ylabel("Popularity",fontsize=20)
plt.title("Popularity of Different parts of Harry Potter",fontsize=25)
plt.show()
```

```python
x = ["day1","day2","day3","day4","day5"]
y = [300,420,250,230,400]
y1 = [500,450,300,250,320]

plt.plot(x,y,marker="p",ls="--",color="red",label="Week1",alpha=0.8)
plt.plot(x,y1,marker="^",ls="-",color="red",label="Week2",alpha=0.5)
plt.legend()
plt.show()
```

```python
x = np.random.randint(10,100,50)
y = np.random.randint(10,100,50)
color = np.random.randint(10,100,50)

plt.scatter(x,y,marker="*",cmap="hot",c=color)
plt.show()
```

```python
brands = ['Oneplus','Apple','Samsung','Nokia','Redmi']
x = [22,35,30,3,10]
c = ['red','Violet','blue','magenta','orange']
ex = [0,0,0,0,0.1]
plt.pie(x,labels=brands,colors=c,explode=ex,shadow=True,autopct="%.2f",startangle=100)
plt.show()
```

```python
l = [25,46,57,25,47,45,13,29,39,46,99,22,12,14,16,55,6,46,52,55,78]
l1 = [13,31,27,56,34,56,77,37,30,20,30,78,45]
plot_values = [l,l1]
plt.boxplot(plot_values)
plt.show()
```

```python
x = [30,40,68,27,44,34,56,13,53,21,53,45,23]
plt.hist(x,bins=15,edgecolor="black",color="pink")
plt.show()
```

```python
x = [10,40,20,40,30,40,60,70,50,40,20,30,40]
plt.stem(x,linefmt = '--',markerfmt="D",orientation = "horizontal")
plt.show()
```

```python
days = [1,2,3,4,5,6,7]

NOP1 = [5,10,30,80,95,90,80]
NOP2 = [10,20,30,75,80,90,120]
NOP3 = [8,30,50,60,70,90,100]

plt.stackplot(days,NOP1,NOP2,NOP3,baseline = "wiggle",colors = ['pink','hotpink','magenta'],labels = ['Week1','Week2','Week3'])
plt.legend()
plt.show()
```

```python
x = ['day1','day2','day3','day4','day5']
y = [30,40,25,30,40]

plt.step(x,y,where='mid',marker='o')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [45,67,33,67,12]
y1 = [41,60,13,66,13]

plt.figure(figsize=[5,3])
plt.plot(x,y,label='male')
plt.plot(x,y1,label='female')
plt.legend(['a1','a2'],ncols = 2,bbox_to_anchor=(0.8,1.2))
plt.show()
```
