```python
dict2 = {
    "Keys": ["k1","k2","k1","k2"],
    "Names": ["John","Ben","David","Peter"],
    "Houses": ["red","blue","green","red"],
    "Grades": ["3rd","8th","9th","8th"]
}

df = pd.DataFrame(dict2)
df.head()
```

```python
df.pivot(index = "Keys",columns = "Names",values = ["Houses","Grades"])
```

```python
pd.melt(df,id_vars=["Names"],value_vars = ["Houses","Grades"],var_name="Houses & Grades",value_name="Values")
```
